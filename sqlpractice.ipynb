# inner join

```python
query = "Select * from customers c inner join order_details o on c.cust_id=o.cust_id"
od = pd.read_sql(query,engine)
od
```

```python
query = '''select c.cust_name, c.cust_id, count(c.cust_id)
from customers c inner join order_details o on c.cust_id=o.cust_id 
group by c.cust_id order by count(c.cust_id) desc;'''
od = pd.read_sql(query,engine)
od
```

```python
query = '''select c.cust_name, c.cust_id, count(c.cust_id)
from customers c left join order_details o on c.cust_id=o.cust_id 
group by c.cust_id order by count(c.cust_id) desc;'''
od = pd.read_sql(query,engine)
od
```

```python
query = '''select c.cust_name, c.cust_id, count(c.cust_id)
from customers c right join order_details o on c.cust_id=o.cust_id 
group by c.cust_id order by count(c.cust_id) desc;'''
od = pd.read_sql(query,engine)
od
```

```python
query = '''select c.cust_name, c.cust_id, count(c.cust_id)
from customers c full outer join order_details o on c.cust_id=o.cust_id 
group by c.cust_id order by count(c.cust_id) desc;'''
od = pd.read_sql(query,engine)
od
```

```python
query = '''select c.cust_name, c.cust_id, count(c.cust_id)
from customers c inner join order_details o on c.cust_id=o.cust_id  and o.qty > 2 
group by c.cust_id order by count(c.cust_id) desc;'''
od = pd.read_sql(query,engine)
od
```

```python
query = "Select * from customers c inner join order_details o on c.cust_id=o.cust_id and o.item_id in (1,2,3 )"
od = pd.read_sql(query,engine)
od
```
